```python
import sys
import os
sys.path.append(os.path.abspath('../../lib'))

import numpy as np
from dhmc import dhmc_transversalization
from dhmc import generate_compositions, capacity, is_dhmc_valid
```

```python
# Example usage:
# A is n x m incidence matrix
# e.g. n=3 elements, m=4 sets
# S0 = {0, 1}
# S1 = {1, 2}
# S2 = {0, 2}
# S3 = {0, 1, 2}
A = np.array([
    [1, 0, 1, 1], # element 0 in S0, S2, S3
    [1, 1, 0, 1], # element 1 in S0, S1, S3
    [0, 1, 1, 1]  # element 2 in S1, S2, S3
])

alpha = [1, 1, 1]

valid_betas = dhmc_transversalization(A, alpha)
print(f"Valid Betas: {valid_betas}")
```
